# **`linear regression`**

**`Linear Regression Definition`:**

Linear regression is a **machine learning method** used to **find the relationship between numbers (data)**.
It helps us **draw a straight line** that can **predict future values** based on given input.

### Example:

If we know:

* Hours studied → 2, 4, 6, 8
* Marks received → 50, 60, 70, 80

Linear regression can learn this pattern and **predict marks** for new hours like 5 or 7.

### In simple words:

✅ It finds the **best straight line** between input and output
✅ It helps us **predict results**
✅ It works only when the relation is roughly **a straight line**

### `Formula`:

[
y = mx + b
]
Where:

* **y** = result (prediction)
* **x** = input (feature)
* **m** = slope of the line
* **b** = intercept (starting value)



# **`Types fo Linear Regression`**

### **Types of Linear Regression:**

There are **two main types**:

---

### **1. Simple Linear Regression**

* Uses **only one input (feature)** to predict output.
* Draws a **straight line**.
* Example:

  * Hours studied → Marks

✅ 1 input → 1 output

---

### **2. Multiple Linear Regression**

* Uses **more than one input (feature)** to predict output.
* Example:

  * Hours studied, Sleep, Attendance → Marks

✅ More than 1 input → 1 output

---

### Short Summary:

| Type                       | Inputs    | Output |
| -------------------------- | --------- | ------ |
| Simple Linear Regression   | 1         | 1      |
| Multiple Linear Regression | 2 or more | 1      |

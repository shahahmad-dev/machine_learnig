````python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from catboost import CatBoostClassifier, Pool, cv
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# load dataset
df = sns.load_dataset('titanic')
````

# **``Preposessing the Data``**

```python

````python
# impute missing values with KNN Imputer in fare, age, embarked and embark_town columns
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
df[['age', 'fare']] = imputer.fit_transform(df[['age', 'fare']])

# impute missing values in 'embarked' and 'embark_town' with simple imputer
from sklearn.impute import SimpleImputer
simple_imputer = SimpleImputer(strategy='most_frequent')
df[['embarked', 'embark_town']] = simple_imputer.fit_transform(df[['embarked', 'embark_town']])

# drop deck column
df.drop('deck', inplace=True, axis=1)
````

````python
df.isnull().sum()
````

````python
# convert categorical columns to 'category' dtype
categorical_cols = df.select_dtypes(include = ['category', 'object'])
for col in categorical_cols:
    df[col] = df[col].astype('category')    
````

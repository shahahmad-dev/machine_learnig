### üîÅ Inverse Encoding / Inverse Transform / Decoding Categorical Variables üòä

**Simple meaning:**
üëâ Converting **encoded numbers back into their original category names**.

---

### üß† Why is it needed?

Machine Learning models work with **numbers**, not text.
So we **encode** categorical data into numbers.
After prediction, to **understand the result**, we convert numbers **back to labels**.

---

### üî¢ Example (Label Encoding)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Encoding
data = ['Cat', 'Dog', 'Fish']
encoded = le.fit_transform(data)
print(encoded)
```

**Output:**

```
[0 1 2]
```

---

### üîÑ Inverse Transform (Decoding)

```python
decoded = le.inverse_transform(encoded)
print(decoded)
```

**Output:**

```
['Cat' 'Dog' 'Fish']
```

---

### ‚úÖ Summary

* **Encoding** ‚Üí Text ‚ûú Numbers üî¢
* **Inverse Transform / Decoding** ‚Üí Numbers ‚ûú Text üî§
* Used to **interpret model predictions** üìä

````python
# load data
df = sns.load_dataset('titanic')
df.head()
````

````python
# encode all categorical variable
le = LabelEncoder()
for col in df.select_dtypes(include=['category', 'object']).columns:
  df[col] = le.fit_transform(df[col])
````

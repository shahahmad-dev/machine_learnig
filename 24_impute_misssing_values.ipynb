```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# load dataset
df = sns.load_dataset('titanic')
```

```python
# check the missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# impute the missing values with mean/mode and median
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
# check the missing values
df.isnull().sum().sort_values(ascending=False)
```

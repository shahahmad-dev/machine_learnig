
## ðŸ”¹ What is Polynomial Regression?

Polynomial Regression is a type of **linear regression** that is used when the relationship between the input (X) and output (y) is **not a straight line**, but a **curve**.

Even though itâ€™s called â€œpolynomial,â€ it is still a **linear model** â€” linear in parameters, not in features.

---

## ðŸ”¹ Formula

**Linear Regression**
[
y = b_0 + b_1x
]

**Polynomial Regression (degree 2)**
[
y = b_0 + b_1x + b_2x^2
]

Higher degree means more curve.

---

## ðŸ”¹ When to Use Polynomial Regression?

* When data shows a **curved pattern**
* When linear regression **underfits**
* When the relationship is **non-linear**

---

## ðŸ”¹ Degree Concept

* Degree 1 â†’ Straight line
* Degree 2 â†’ Parabola (U-shape)
* Degree 3 â†’ More complex curve

âš ï¸ Higher degree can cause **overfitting**.

---

## ðŸ”¹ Real-World Examples

* Salary vs experience
* Speed vs braking distance
* Temperature vs electricity usage

## ðŸ”¹ Polynomial vs Linear Regression

| Feature     | Linear        | Polynomial  |
| ----------- | ------------- | ----------- |
| Shape       | Straight line | Curve       |
| Complexity  | Low           | Mediumâ€“High |
| Overfitting | Low           | High        |

---

### ðŸ”¥ Interview Tip

> Polynomial regression is **linear in coefficients but non-linear in features**.



````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# create dummy data
np.random.seed(0)
x = 2-3*np.random.normal(0,1,100)
y = x**2 - 1 + 0.5*np.random.normal(0,1,100)

# reshape
x = x.reshape(-1,1)
y = y.reshape(-1,1)

# polynomial transformation
degree = 2
poly = PolynomialFeatures(degree=degree)
x_poly = poly.fit_transform(x)

# linear regression
model = LinearRegression()
model.fit(x_poly, y)

y_pred = model.predict(x_poly)

# plot the result using plotly
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=x.flatten(), y=y.flatten(), mode='markers', name='data'))
fig.add_trace(go.Scatter(x=x.flatten(), y=y_pred.flatten(), mode='lines', name='linear regression'))
fig.show()
````
